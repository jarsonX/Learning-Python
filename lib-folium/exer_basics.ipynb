```python
import folium
```

```python
# Canada

canada_map = folium.Map(location=[56.130, -106.35], zoom_start=4)

ontario = folium.map.FeatureGroup()

ontario.add_child(folium.features.CircleMarker([51.25, -85.32], radius=5, color='red', fill_color='red'))
canada_map.add_child(ontario)
folium.Marker([51.25, -85.32], popup='Ontario').add_to(canada_map)

canada_map
```

```python
# Poland

poland_map = folium.Map(location=[51, 20], zoom_start=6)

katowice = folium.map.FeatureGroup()

katowice.add_child(folium.features.CircleMarker([50.264340516219946, 19.02356493256628], radius=10, color='red', fill_color='white'))
poland_map.add_child(katowice)
folium.Marker([50.264340516219946, 19.02356493256628], popup='Katowice').add_to(poland_map)

poland_map
```
